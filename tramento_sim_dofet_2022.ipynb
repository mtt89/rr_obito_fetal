```python
# ocor_MUNNOMEX,res_MUNNOMEX,ocor_CAPITAL,res_CAPITAL,ocor_REGIAO,res_REGIAO,ocor_SIGLA_UF,res_SIGLA_UF
sim_dofet_2019 = pd.read_csv('SIM_DOFET\\ETLSIM.DOFET_2019_t.csv')
sim_dofet_2020 = pd.read_csv('SIM_DOFET\\ETLSIM.DOFET_2020_t.csv')
sim_dofet_2021 = pd.read_csv('SIM_DOFET\\ETLSIM.DOFET_2021_t.csv')

sim_dofet_unif = pd.concat([sim_dofet_2019, sim_dofet_2020, sim_dofet_2021])
```

```python
sim_dofet_res = sim_dofet_unif[
    [
        'CODMUNRES'
        , 'res_MUNNOMEX'
        , 'res_CAPITAL'
        , 'res_REGIAO'
        , 'res_SIGLA_UF'
    ]
].drop_duplicates()

sim_dofet_ocor = sim_dofet_unif[
    [
        'CODMUNOCOR'
        , 'ocor_MUNNOMEX'
        , 'ocor_CAPITAL'
        , 'ocor_REGIAO' 
        , 'ocor_SIGLA_UF'
    ]
].drop_duplicates()
```

```python
df_2022_merge = df_2022.merge(sim_dofet_ocor, how='left', on='CODMUNOCOR')
```

```python
df_2022_merge = df_2022_merge.merge(sim_dofet_res, how='left', on='CODMUNRES')
```

```python
df_2022_merge.info()
```

```python
sim_dofet_causa = sim_dofet_unif[
    [
        'CAUSABAS'
        , 'causabas_capitulo'
        , 'causabas_categoria'
        , 'causabas_grupo'
        , 'causabas_subcategoria'
                ]
].drop_duplicates()
```

```python
len(sim_dofet_causa)
```

```python
df_2022_merge = df_2022_merge.merge(sim_dofet_causa, how='left', on='CAUSABAS')
```

```python
df_2022_merge['FLAG_BASE'] = 'SIM_DOFET'
```

```python
def acerta_data(string):
    tam = len(string)
    if tam ==8:
        dt = f'{string[4:8]}-{string[2:4]}-{string[0:2]}'
        return dt
    elif tam == 7:
        dt = f'{string[3:7]}-{string[1:3]}-0{string[0]}'
        return dt
    else:
        None
        
    
df_2022_merge['data_obito'] = [acerta_data(str(i)) for i in df_2022_merge['DTOBITO']]
```

```python
df_2022_merge['ano_obito'] = [i[0:4] for i in df_2022_merge['data_obito']]
```

```python
def escolaridade_mae(cod):
    if cod == 0:
        return 'Sem_escolaridade'
    elif 1 <= cod <= 2:
        return 'Fundamental'
    elif cod == 3:
        return 'Ensino_medio'
    elif 4 <= cod <= 5:
        return 'Ensino_superior'
    else:
        return 'Ignorado'

df_2022_merge['escolaridade_mae'] = [escolaridade_mae(cod) for cod in df_2022_merge['ESCMAE2010']]
```

```python
def sexo(cod):
    if cod == 1:
        return 'Masculino'
    elif cod == 2:
        return 'Feminino'
    else:
         return 'Ignorado'
df_2022_merge['def_sexo'] = [sexo(cod) for cod in df_2022_merge['SEXO']]
```

```python
def obito_parto(cod):
    if cod == 1:
        return 'Antes'
    elif cod == 2:
        return 'Durante'
    elif cod == 3:
        return 'Depois'
    else:
         return 'Ignorado'
df_2022_merge['def_obito_parto'] = [obito_parto(cod) for cod in df_2022_merge['OBITOPARTO']]
```

```python
def estado_civil(cod):
    if cod == 1:
        return 'Solteiro'
    elif cod == 2:
        return 'Casado'
    elif cod == 3:
        return 'Viuvo'
    elif cod == 4:
        return 'Separado_judicialmente'
    elif cod == 5:
        return 'Uniao_consensual'
    else:
         return 'Ignorado'
df_2022_merge['def_est_civil'] = [estado_civil(i) for i in df_2022_merge['ESTCIV']]
```

```python
def raca_cor(cod):
    if cod == 1:
        return 'Branca'
    elif cod == 2:
        return 'Preta'
    elif cod == 3:
        return 'Amarela'
    elif cod == 4:
        return 'Parda'
    elif cod == 5:
        return 'Indigena'
    else:
         return 'Ignorado'
df_2022_merge['def_raca_cor'] = [raca_cor(i) for i in df_2022_merge['RACACOR']]
```

```python
def qtd_filhos_mortos(cod):
    if cod == 0:
        return 'sem_filhos'
    elif cod == 1.0:
        return '1_filho'
    elif (cod >= 2.0) & (cod <= 4.0):
        return '2_a_4_filhos'
    elif (cod > 4.0) & (cod <= 98.0):
        return '5_ou_mais_filhos'
    else:
         return 'Ignorado'
df_2022_merge['def_num_filhos_mortos'] = [qtd_filhos_mortos(i) for i in df_2022_merge['QTDFILMORT']]
```
