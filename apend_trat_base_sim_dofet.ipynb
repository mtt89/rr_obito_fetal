```python
def obito_parto(cod):
    if cod == 1:
        return 'Antes'
    elif cod == 2:
        return 'Durante'
    elif cod == 3:
        return 'Depois'
    else:
         return 'Ignorado'

def estado_civil(cod):
    if cod == 1:
        return 'Solteiro'
    elif cod == 2:
        return 'Casado'
    elif cod == 3:
        return 'Viuvo'
    elif cod == 4:
        return 'Separado_judicialmente'
    elif cod == 5:
        return 'Uniao_consensual'
    else:
         return 'Ignorado'
        
def raca_cor(cod):
    if cod == 1:
        return 'Branca'
    elif cod == 2:
        return 'Preta'
    elif cod == 3:
        return 'Amarela'
    elif cod == 4:
        return 'Parda'
    elif cod == 5:
        return 'Indigena'
    else:
         return 'Ignorado'

def paridade(cod):
    if cod == 0.0:
        return 'Primiparous'
    elif (cod > 0.0) & (cod <= 98.0):
        return 'Multiparous'
    else:
         return 'Ignorado'

def func_categorize_gravidez(cod_grav):
    if cod_grav == 1:
        return "Unica"
    elif 2 <= cod_grav <= 3:
        return "Multipla"
    else:
        return "Ignorado"

def func_obito_parto(cod):
    if cod == 1:
        return "Antes"
    elif cod == 2:
        return "Durante"
    elif cod == 3:
        return "Depois"
    else:
        return "Ignorado"
        
def func_parto(cod):
    if cod == 1:
        return "Vaginal"
    elif cod == 2:
        return "Cesareo"
    else:
        return "Ignorado"

def safe_peso_calc(sexo, peso, semana):
    # Se algum argumento essencial for nulo, retorna np.nan
    if pd.isna(sexo) or pd.isna(peso) or pd.isna(semana):
        return np.nan
    # Caso contrário, chama a função original
    return func_peso_calculado(sexo, peso, int(round(semana)))
```

```python
# Outliers
def detectar_outliers(col):
    q1 = col.quantile(0.25)
    q3 = col.quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers = col[(col < lower_bound) | (col > upper_bound)]
    return outliers

outliers_numericos = df_sim_dofet[lista_numerica].apply(detectar_outliers)

for i in lista_numerica:
    print(outliers_numericos.loc[~outliers_numericos[i].isna(), [i]].describe())
```
